````python
import getpass
import os
import sqlite3
from langchain_openai import ChatOpenAI
from langchain_community.utilities import SQLDatabase

# Step 1: Set API Keys (if not already set in your environment)
if not os.environ.get("OPENAI_API_KEY"):
    os.environ["OPENAI_API_KEY"] = getpass.getpass("Enter your OpenAI API Key: ")

# Step 2: Connect to SQL Database
db_path = "test_database.db"
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# Sample table and data creation
cursor.execute('''CREATE TABLE IF NOT EXISTS employees (
                    id INTEGER PRIMARY KEY,
                    name TEXT,
                    age INTEGER,
                    department TEXT
                )''')
cursor.executemany('INSERT OR IGNORE INTO employees (id, name, age, department) VALUES (?, ?, ?, ?)', [
    (1, 'Alice', 30, 'HR'),
    (2, 'Bob', 24, 'Engineering'),
    (3, 'Charlie', 28, 'Sales'),
])
conn.commit()
conn.close()

# Step 3: Set Up LangChain SQLDatabase
db = SQLDatabase.from_uri(f"sqlite:///{db_path}")
llm = ChatOpenAI(model_name="gpt-3.5-turbo", api_key=os.getenv("OPENAI_API_KEY"))

# Step 4: Define a function to convert natural language to SQL
def natural_language_to_sql(natural_language_query):
    # Here we use the language model to generate an SQL query from the natural language input
    prompt = f"Convert the following question into a SQL query: {natural_language_query}"
    sql_query = llm.invoke([{"role": "user", "content": prompt}])  # Use `invoke` method
    return sql_query['content']  # Extract the content from the response

# Step 5: Define a function to execute SQL queries
def execute_query(query):
    try:
        # Connect to the database
        conn = sqlite3.connect(db_path)
        cursor = conn.cursor()
        
        # Execute the query
        cursor.execute(query)
        results = cursor.fetchall()
        
        # Commit if it's an insert/update/delete operation
        conn.commit()
        
        # Close the connection
        conn.close()
        
        return results
    except Exception as e:
        print("Error executing query:", e)
        return None

# Step 6: Define a function to test natural language queries
def query_database(natural_language_query):
    sql_query = natural_language_to_sql(natural_language_query)
    print("Generated SQL Query:", sql_query)  # Show the generated SQL query for debugging
    response = execute_query(sql_query)
    if response is not None:
        print("Response:", response)
    else:
        print("No response or an error occurred.")

# Step 7: Test with a few queries
print("Ask questions about the database, like 'How many employees are in HR?'")
while True:
    user_query = input("Enter your question (or type 'exit' to quit): ")
    if user_query.lower() == 'exit':
        break
    query_database(user_query)
````

````python
import getpass
import os
import requests
import psycopg2
import matplotlib.pyplot as plt
from langchain_community.utilities import SQLDatabase

# Step 1: Configurar chaves de API do Azure OpenAI e o Endpoint
if not os.environ.get("AZURE_OPENAI_API_KEY"):
    os.environ["AZURE_OPENAI_API_KEY"] = getpass.getpass("Digite sua chave de API do Azure OpenAI: ")
if not os.environ.get("AZURE_OPENAI_ENDPOINT"):
    os.environ["AZURE_OPENAI_ENDPOINT"] = input("Digite seu endpoint do Azure OpenAI completo: ")

# Configurações específicas para a API do Azure OpenAI
azure_api_key = os.environ["AZURE_OPENAI_API_KEY"]
endpoint_url = os.environ["AZURE_OPENAI_ENDPOINT"]

# Configuração de conexão com PostgreSQL
postgres_user = "postgres"
postgres_password = "cea-upbi"
postgres_host = "195.200.4.223"
postgres_dbname = "upbi"
postgres_port = 5432

conn_str = f"postgresql://{postgres_user}:{postgres_password}@{postgres_host}:{postgres_port}/{postgres_dbname}"
db = SQLDatabase.from_uri(conn_str)

# Função para listar a estrutura do banco de dados
def get_database_schema():
    conn = psycopg2.connect(
        dbname=postgres_dbname,
        user=postgres_user,
        password=postgres_password,
        host=postgres_host,
        port=postgres_port
    )
    cursor = conn.cursor()
    schema = {}

    cursor.execute("SELECT table_name FROM information_schema.tables WHERE table_schema='public';")
    tables = cursor.fetchall()
    for table_name in tables:
        table_name = table_name[0]
        cursor.execute(f"SELECT column_name FROM information_schema.columns WHERE table_name = '{table_name}';")
        columns = cursor.fetchall()
        schema[table_name] = [column[0] for column in columns]  # Nome das colunas

    conn.close()
    return schema

# Função para converter linguagem natural para SQL e código de gráfico
def natural_language_to_sql_and_plot(natural_language_query):
    schema = get_database_schema()
    schema_description = "\n".join([f"Tabela {table}: {columns}" for table, columns in schema.items()])
    
    prompt = (
        f"Com base na estrutura do banco de dados abaixo, responda à pergunta '{natural_language_query}' "
        f"usando uma consulta SQL apropriada e gere o código Python para plotar um gráfico relevante com Matplotlib.\n\n"
        f"Estrutura do banco de dados:\n{schema_description}\n\n"
        "Formato da resposta:\n\n"
        "SQL Query:\n"
        "```sql\nSELECT ... FROM ... WHERE ...;\n```\n\n"
        "Python code for plot:\n"
        "```python\nimport matplotlib.pyplot as plt\n# Código para gerar o gráfico\nplt.show()\n```\n\n"
        "Apenas a consulta SQL e o código de plotagem:"
    )

    headers = {
        "Content-Type": "application/json",
        "api-key": azure_api_key
    }
    data = {
        "messages": [
            {"role": "system", "content": "Você é um assistente SQL que também sabe criar gráficos."},
            {"role": "user", "content": prompt}
        ],
        "temperature": 0,
        "max_tokens": 1500,
        "top_p": 1,
        "frequency_penalty": 0,
        "presence_penalty": 0
    }

    response = requests.post(endpoint_url, headers=headers, json=data)
    response.raise_for_status()

    response_data = response.json()
    response_text = response_data['choices'][0]['message']['content'].strip()

    # Verificar se a resposta contém a seção "Python code for plot:"
    if "Python code for plot:" in response_text:
        sql_query, plot_code = response_text.split("Python code for plot:")
        
        # Limpeza da consulta SQL
        sql_query = sql_query.replace("SQL Query:", "").strip()  # Remove o prefixo "SQL Query:"
        sql_query = sql_query.replace("```sql", "").replace("```", "").strip()  # Remove delimitadores de código
        plot_code = plot_code.strip()
        
        return sql_query, plot_code
    else:
        raise ValueError("A resposta não contém o código de plotagem esperado.")

# Função para executar consultas SQL
def execute_query(query):
    try:
        conn = psycopg2.connect(
            dbname=postgres_dbname,
            user=postgres_user,
            password=postgres_password,
            host=postgres_host,
            port=postgres_port
        )
        cursor = conn.cursor()
        cursor.execute(query)
        results = cursor.fetchall()
        conn.commit()
        conn.close()
        return results
    except Exception as e:
        print("Erro ao executar a consulta:", e)
        return None

# Função para gerar gráfico de barras com dados de consulta
def generate_bar_chart(data, x_label, y_label, title):
    items = [row[0] for row in data]
    values = [row[1] for row in data]

    plt.figure(figsize=(10, 5))
    plt.bar(items, values, color='skyblue')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Função para testar consultas e gráficos
def query_database_with_chart(natural_language_query):
    try:
        sql_query, _ = natural_language_to_sql_and_plot(natural_language_query)
        print("Consulta SQL Gerada:", sql_query)
        
        # Executar a consulta SQL
        response = execute_query(sql_query)
        
        if response is not None:
            print("Resposta da Consulta:", response)
            
            # Perguntar se o usuário deseja visualizar o gráfico
            user_choice = input("Deseja gerar um gráfico com os dados retornados? (s/n): ")
            if user_choice.lower() == 's':
                # Defina rótulos e título para o gráfico
                x_label = "Itens"
                y_label = "Total de Vendas"
                title = "Total de Vendas por Item"
                generate_bar_chart(response, x_label, y_label, title)
        else:
            print("Nenhuma resposta ou ocorreu um erro.")
    except ValueError as ve:
        print("Erro:", ve)
    except Exception as e:
        print("Erro inesperado:", e)

# Teste com consultas em português para gerar gráficos
print("Faça perguntas sobre o banco de dados, como 'Quantos itens ativos existem no sistema?' ou 'Mostre um gráfico de vendas.'")
while True:
    user_query = input("Digite sua pergunta (ou digite 'sair' para encerrar): ")
    if user_query.lower() == 'sair':
        break
    query_database_with_chart(user_query)
````
